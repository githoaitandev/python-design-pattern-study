# Single Responsibility Principle (Python)

**Mục tiêu:** hiểu rằng một class chỉ nên có một lý do để thay đổi.

**Ý chính:** nếu một class vừa chứa business logic, vừa lo lưu file, vừa lo hiển thị dữ liệu, thì class đó đang gánh quá nhiều trách nhiệm.


## Before: vi phạm Single Responsibility Principle

`Report` vừa quản lý dữ liệu báo cáo, vừa format nội dung, vừa ghi file ra đĩa.


```python
class Report:
    def __init__(self, title: str, content: str):
        self.title = title
        self.content = content

    def format(self) -> str:
        return f"Title: {self.title}\nContent: {self.content}"

    def save_to_file(self, filename: str):
        with open(filename, "w", encoding="utf-8") as file:
            file.write(self.format())

    def print_report(self):
        print(self.format())
```

### Vấn đề

- Nếu thay đổi cách format, class này phải sửa.
- Nếu thay đổi cách lưu file, class này cũng phải sửa.
- Nếu thêm cách xuất khác như database hoặc API, class tiếp tục phình to.


## After: áp dụng đúng Single Responsibility Principle

Tách mỗi trách nhiệm ra một class riêng: dữ liệu báo cáo, format báo cáo, và lưu báo cáo.


```python
class Report:
    def __init__(self, title: str, content: str):
        self.title = title
        self.content = content


class ReportFormatter:
    def format(self, report: Report) -> str:
        return f"Title: {report.title}\nContent: {report.content}"


class ReportFileSaver:
    def save(self, report: Report, formatter: ReportFormatter, filename: str):
        with open(filename, "w", encoding="utf-8") as file:
            file.write(formatter.format(report))


class ReportPrinter:
    def print(self, report: Report, formatter: ReportFormatter):
        print(formatter.format(report))
```

```python
report = Report("Weekly Summary", "Completed SOLID examples")
formatter = ReportFormatter()
printer = ReportPrinter()

printer.print(report, formatter)
```

### Điểm đúng ở bản sau

- `Report` chỉ giữ dữ liệu.
- `ReportFormatter` chỉ lo format.
- `ReportFileSaver` chỉ lo lưu file.
- Mỗi class chỉ có một lý do để thay đổi.
