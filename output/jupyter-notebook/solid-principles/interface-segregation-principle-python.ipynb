# Interface Segregation Principle (Python)

**Mục tiêu:** hiểu rằng client không nên bị ép phụ thuộc vào method mà nó không dùng.

**Ý chính:** interface lớn, ôm đồm thường làm các class triển khai phải gánh những method không phù hợp.


## Before: vi phạm Interface Segregation Principle

`Worker` có cả `work()` và `eat()`, nhưng không phải mọi worker đều cần `eat()`.


```python
from abc import ABC, abstractmethod


class Worker(ABC):
    @abstractmethod
    def work(self):
        raise NotImplementedError

    @abstractmethod
    def eat(self):
        raise NotImplementedError


class HumanWorker(Worker):
    def work(self):
        print("Human is working")

    def eat(self):
        print("Human is eating")


class RobotWorker(Worker):
    def work(self):
        print("Robot is working")

    def eat(self):
        raise NotImplementedError("Robot does not eat")
```

### Vấn đề

- `RobotWorker` bị ép triển khai `eat()` dù không phù hợp.
- Interface đang quá to và không phản ánh đúng nhu cầu của từng client.


## After: áp dụng đúng Interface Segregation Principle

Tách interface lớn thành các interface nhỏ hơn, đúng theo hành vi thực sự cần dùng.


```python
class Workable(ABC):
    @abstractmethod
    def work(self):
        raise NotImplementedError


class Eatable(ABC):
    @abstractmethod
    def eat(self):
        raise NotImplementedError


class HumanWorker(Workable, Eatable):
    def work(self):
        print("Human is working")

    def eat(self):
        print("Human is eating")


class RobotWorker(Workable):
    def work(self):
        print("Robot is working")
```

```python
human = HumanWorker()
robot = RobotWorker()

human.work()
human.eat()
robot.work()
```

### Điểm đúng ở bản sau

- `RobotWorker` không còn phải triển khai method vô nghĩa.
- Client nào cần `work()` chỉ phụ thuộc vào `Workable`.
- Client nào cần `eat()` thì phụ thuộc vào `Eatable`.
