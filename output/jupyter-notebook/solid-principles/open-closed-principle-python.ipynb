# Open-Closed Principle (Python)

**Mục tiêu:** hiểu rằng code nên mở để mở rộng, nhưng đóng để chỉnh sửa.

**Ý chính:** khi có yêu cầu mới, ta nên thêm code mới thay vì sửa liên tục code cũ đã ổn định.


## Before: vi phạm Open-Closed Principle

Class `DiscountCalculator` dùng `if/elif` để xử lý từng loại khách hàng. Mỗi khi có loại mới, ta phải sửa code cũ.


```python
class DiscountCalculator:
    def calculate(self, customer_type: str, amount: float) -> float:
        if customer_type == "regular":
            return amount * 0.05
        elif customer_type == "vip":
            return amount * 0.10
        elif customer_type == "premium":
            return amount * 0.15
        else:
            return 0.0
```

### Vấn đề

- Mỗi lần thêm loại khách hàng mới phải sửa class đang chạy ổn định.
- Dễ phát sinh bug ở logic cũ.
- Code ngày càng dài vì `if/elif`.


## After: áp dụng đúng Open-Closed Principle

Tạo abstraction cho chiến lược giảm giá, rồi thêm class mới khi cần mở rộng.


```python
from abc import ABC, abstractmethod


class DiscountStrategy(ABC):
    @abstractmethod
    def calculate(self, amount: float) -> float:
        raise NotImplementedError


class RegularDiscount(DiscountStrategy):
    def calculate(self, amount: float) -> float:
        return amount * 0.05


class VipDiscount(DiscountStrategy):
    def calculate(self, amount: float) -> float:
        return amount * 0.10


class PremiumDiscount(DiscountStrategy):
    def calculate(self, amount: float) -> float:
        return amount * 0.15


class DiscountCalculator:
    def calculate(self, strategy: DiscountStrategy, amount: float) -> float:
        return strategy.calculate(amount)
```

```python
calculator = DiscountCalculator()

print(calculator.calculate(RegularDiscount(), 1000))
print(calculator.calculate(VipDiscount(), 1000))
print(calculator.calculate(PremiumDiscount(), 1000))
```

- Không cần sửa logic cũ trong `DiscountCalculator`.
- Hệ thống dễ mở rộng hơn và ít rủi ro hơn.
