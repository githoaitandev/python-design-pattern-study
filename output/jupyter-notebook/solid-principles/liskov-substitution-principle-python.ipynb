# Liskov Substitution Principle (Python)

**Mục tiêu:** hiểu vì sao một lớp con phải thay thế được lớp cha mà không làm thay đổi tính đúng đắn của chương trình.

**Ý chính:** nếu code đang hoạt động đúng với kiểu cha, thì khi truyền vào kiểu con, hành vi cốt lõi vẫn phải đúng theo kỳ vọng đó.


## Before: vi phạm Liskov Substitution Principle

Ví dụ kinh điển là `Rectangle` và `Square`.

Hàm client kỳ vọng rằng có thể đặt `width` và `height` độc lập. Nhưng `Square` lại ghi đè hành vi để hai giá trị luôn bằng nhau, làm phá vỡ kỳ vọng của lớp cha.


```python
class Rectangle:
    def __init__(self, width: int, height: int):
        self.width = width
        self.height = height

    def set_width(self, width: int):
        self.width = width

    def set_height(self, height: int):
        self.height = height

    def area(self) -> int:
        return self.width * self.height


class Square(Rectangle):
    def set_width(self, width: int):
        self.width = width
        self.height = width

    def set_height(self, height: int):
        self.width = height
        self.height = height


def resize_and_print_area(rect: Rectangle):
    rect.set_width(5)
    rect.set_height(10)
    print(f"Expected area = 50, actual area = {rect.area()}")
```

```python
rectangle = Rectangle(2, 3)
square = Square(2, 2)

resize_and_print_area(rectangle)
resize_and_print_area(square)
```

### Vấn đề

- `resize_and_print_area()` mong đợi mọi `Rectangle` đều có thể thay đổi `width` và `height` riêng biệt.
- `Square` không giữ đúng hợp đồng hành vi đó.
- Vì vậy `Square` không thể thay thế an toàn cho `Rectangle`.


## After: áp dụng đúng Liskov Substitution Principle

Cách sửa là không ép `Square` kế thừa `Rectangle` nếu hành vi không thật sự tương thích.

Thay vào đó, để cả hai cùng triển khai một abstraction phù hợp hơn là `Shape`.


```python
from abc import ABC, abstractmethod


class Shape(ABC):
    @abstractmethod
    def area(self) -> int:
        raise NotImplementedError


class BetterRectangle(Shape):
    def __init__(self, width: int, height: int):
        self.width = width
        self.height = height

    def area(self) -> int:
        return self.width * self.height


class BetterSquare(Shape):
    def __init__(self, side: int):
        self.side = side

    def area(self) -> int:
        return self.side * self.side


def print_area(shape: Shape):
    print(f"Area = {shape.area()}")
```

```python
rectangle = BetterRectangle(5, 10)
square = BetterSquare(5)

print_area(rectangle)
print_area(square)
```

### Điểm đúng ở bản sau

- `BetterRectangle` và `BetterSquare` đều thay thế được cho `Shape`.
- Không có lớp con nào phá vỡ kỳ vọng hành vi của abstraction chung.
- Kế thừa chỉ được dùng khi lớp con giữ đúng hợp đồng của lớp cha.


## Ghi nhớ nhanh

- Quan hệ ngoài đời thật chưa đủ để quyết định kế thừa trong code.
- Hãy ưu tiên hành vi và hợp đồng của abstraction.
- Nếu lớp con phải “bẻ” logic của lớp cha để hoạt động, đó thường là dấu hiệu vi phạm LSP.
