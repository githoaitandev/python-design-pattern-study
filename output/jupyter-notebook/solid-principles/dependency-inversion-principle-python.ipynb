# Dependency Inversion Principle (Python)

**Mục tiêu:** hiểu rằng module cấp cao không nên phụ thuộc trực tiếp vào module cấp thấp.

**Ý chính:** cả hai nên phụ thuộc vào abstraction, để code dễ thay thế và dễ test hơn.


## Before: vi phạm Dependency Inversion Principle

`OrderService` phụ thuộc trực tiếp vào `MySQLDatabase`.


```python
class MySQLDatabase:
    def save(self, data: str):
        print(f"Saving '{data}' to MySQL")


class OrderService:
    def __init__(self):
        self.database = MySQLDatabase()

    def create_order(self, order_data: str):
        self.database.save(order_data)
```

```python
service = OrderService()
service.create_order("order-001")
```

### Vấn đề

- `OrderService` bị khóa cứng vào `MySQLDatabase`.
- Nếu đổi sang PostgreSQL, file, API, hoặc mock để test thì phải sửa class cấp cao.


## After: áp dụng đúng Dependency Inversion Principle

Đưa phụ thuộc về abstraction, rồi inject implementation cụ thể từ bên ngoài.


```python
from abc import ABC, abstractmethod


class Database(ABC):
    @abstractmethod
    def save(self, data: str):
        raise NotImplementedError


class MySQLDatabase(Database):
    def save(self, data: str):
        print(f"Saving '{data}' to MySQL")


class FileDatabase(Database):
    def save(self, data: str):
        print(f"Saving '{data}' to file")


class OrderService:
    def __init__(self, database: Database):
        self.database = database

    def create_order(self, order_data: str):
        self.database.save(order_data)
```

```python
mysql_service = OrderService(MySQLDatabase())
file_service = OrderService(FileDatabase())

mysql_service.create_order("order-001")
file_service.create_order("order-002")
```

### Điểm đúng ở bản sau

- `OrderService` phụ thuộc vào `Database`, không phụ thuộc vào class cụ thể.
- Có thể thay implementation mà không phải sửa logic cấp cao.
- Việc test cũng dễ hơn vì có thể truyền mock hoặc fake implementation.
