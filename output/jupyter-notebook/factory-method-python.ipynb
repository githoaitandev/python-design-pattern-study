# Factory Method (Python)

**Mục tiêu:** hiểu *Factory Method* là gì, dùng khi nào, và viết được ví dụ OOP cơ bản.

**Ý chính:** định nghĩa một method tạo đối tượng (factory method) trong lớp cha (Creator), nhưng để lớp con quyết định tạo **đối tượng cụ thể** nào.


## Khi nào dùng?

- Bạn có logic xử lý “cố định” ở lớp cha, nhưng việc tạo đối tượng phụ thuộc vào biến thể.
- Bạn muốn tránh `if/elif` tạo object rải rác ở nhiều nơi.

So sánh nhanh:
- **Factory Method:** tạo *một loại sản phẩm* (theo interface), chọn concrete class bằng cách override.
- **Abstract Factory:** tạo *một họ sản phẩm* (nhiều loại) theo từng biến thể.


```python
from __future__ import annotations

from abc import ABC, abstractmethod
```

## Ví dụ: Logistics tạo phương tiện vận chuyển

Bài toán:
- Ứng dụng có flow xử lý đơn hàng giống nhau.
- Nhưng cách vận chuyển khác nhau theo loại logistics: đường bộ (truck) hoặc đường biển (ship).

Ta sẽ:
1) Định nghĩa `Transport` (Product interface)
2) Định nghĩa `Logistics` (Creator) với `create_transport()` là **factory method**
3) Các lớp con override `create_transport()` để tạo `Truck` hoặc `Ship`


## 1) Product interface


```python
class Transport(ABC):
    @abstractmethod
    def deliver(self) -> str:
        raise NotImplementedError
```

## 2) Concrete products


```python
class Truck(Transport):
    def deliver(self) -> str:
        return "Deliver by road (truck)"


class Ship(Transport):
    def deliver(self) -> str:
        return "Deliver by sea (ship)"
```

## 3) Creator + Factory Method


```python
class Logistics(ABC):
    # Đây là phần logic "ổn định" dùng chung
    def plan_delivery(self) -> str:
        transport = self.create_transport()  # gọi factory method
        return f"Plan: {transport.deliver()}"

    # Factory Method: lớp con quyết định tạo concrete product nào
    @abstractmethod
    def create_transport(self) -> Transport:
        raise NotImplementedError


class RoadLogistics(Logistics):
    def create_transport(self) -> Transport:
        return Truck()


class SeaLogistics(Logistics):
    def create_transport(self) -> Transport:
        return Ship()
```

## 4) Client code

Client chỉ làm việc với `Logistics` (Creator). Đổi biến thể = đổi subclass.


```python
def run(logistics: Logistics) -> str:
    return logistics.plan_delivery()


run(RoadLogistics()), run(SeaLogistics())
```

## Ghi chú / Pitfalls

- Nếu bạn chỉ cần tạo object đơn giản ở 1 chỗ, constructor thường đủ.
- Nếu bạn thấy mình override factory method chỉ để return 1 class mà không có biến thể thực sự, có thể đang over-engineer.
- Factory Method hay đi kèm Template Method: `plan_delivery()` là khung cố định, còn `create_transport()` là điểm mở rộng.
